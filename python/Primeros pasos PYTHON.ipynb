#### Ciclo WHILE 

```python
import numpy as np
```

```python
np.random.random()
```

```python
contador = 0
total = 0


while contador < 100: 
    valor1 = np.random.random()
    valor2 = np.random.random() - 0.1
    if valor1 < valor2:
        contador = contador + 1
    total = total + 1
```

### Funciones

```python
#Calcular el interes compuesto, dado una tasa, un monto inicial y un tiempo
rate = 0.01
amount = 10000
time = 48

ic  = amount*(1+rate)**time
print(ic)

def ic (rate, amount, time):
    return amount*(1+rate)**time
```

```python
ic(0.04, 10000,12)
```

```python
rates = [0.032,0.1205,0.38,0.4035,0.46, 0.925, 0.97]
values = []
for r in rates:
    v = (ic(r,10000,12))
    print (v)
    values.append(v)
```
